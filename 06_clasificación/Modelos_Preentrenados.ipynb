```python
from transformers import pipeline
```

```python
sentiment_pipeline = pipeline("sentiment-analysis")
```

```python
data = ["I love you", "I hate you", "Te amo", "Te odio", "Dame más!"]
sentiment_pipeline(data)
```

```python
# Podemos especificar otro modelo preentrenado 
# https://huggingface.co/finiteautomata
specific_model = pipeline(model="finiteautomata/bertweet-base-sentiment-analysis")
```

```python
specific_model(data)
```

```python
specific_model_2 = pipeline(model="finiteautomata/bert-non-contextualized-hate-speech-es")
specific_model_2(data)
```
