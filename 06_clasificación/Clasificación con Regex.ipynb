```python
from google.colab import drive
import json

drive.mount('/content/drive', force_remount=True)
with open("/content/drive/My Drive/twitter.json", 'r') as f:
    api_token= json.load(f)
```

Tweepy

```python
import tweepy as tw
# your Twitter API key and API secret
my_api_key = api_token["APIKey"]
my_api_secret = api_token["APIKeySecret"]
# authenticate
auth = tw.OAuthHandler(my_api_key, my_api_secret)
api = tw.API(auth, wait_on_rate_limit=True)
```

```python
search_query = "covid"

tweets = tw.Cursor(api.search,
              q=search_query,
              lang="es",
              since="2021-02-16").items(50)
# store the API responses in a list
tweets_copy = []
for tweet in tweets:
    tweets_copy.append(tweet)
    
print("Total Tweets fetched:", len(tweets_copy))
```

```python
import pandas as pd
# intialize the dataframe
tweets_df = pd.DataFrame()
# populate the dataframe
for tweet in tweets_copy:
    hashtags = []
    try:
        for hashtag in tweet.entities["hashtags"]:
            hashtags.append(hashtag["text"])
        text = api.get_status(id=tweet.id, tweet_mode='extended').full_text
    except:
        pass
    tweets_df = tweets_df.append(pd.DataFrame({'user_name': tweet.user.name, 
                                               'user_location': tweet.user.location,\
                                               'user_description': tweet.user.description,
                                               'user_verified': tweet.user.verified,
                                               'date': tweet.created_at,
                                               'text': text, 
                                               'hashtags': [hashtags if hashtags else None],
                                               'source': tweet.source}))
    tweets_df = tweets_df.reset_index(drop=True)
# show the dataframe
tweets_df.head()
```

```python
pattern = r'@\S+'
```

```python
import re
def arroba(tweet: str): 
  return bool(re.search(pattern, tweet))
```

```python
tweets_df['arroba'] = tweets_df['text'].apply(lambda x: arroba(x))
```
