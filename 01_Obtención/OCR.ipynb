```python
import pytesseract
import cv2
from os import listdir
from datetime import date, timedelta
```

```python
pytesseract.pytesseract.tesseract_cmd = 'C:/Program Files/Tesseract-OCR/tesseract.exe'  # your path may be different
```

```python
def daterange(start_date, end_date):
    for n in range(int((end_date - start_date).days)):
        yield start_date + timedelta(n)
```

```python
custom_config = r'--oem 3 --psm 6'
a=""


start_date = date(2010, 1, 1)
end_date = date(2021, 6, 20)

#for file in listdir("c:/data/tapas/"):
for sd in daterange(start_date, end_date):    
    file=sd.strftime("%Y")+sd.strftime("%m")+sd.strftime("%d")+".jpg"
    img = cv2.imread("c:/data/tapas/" + file)
    try:
        nuevo=pytesseract.image_to_string(img, config=custom_config).replace("\n", "").replace(";",",")
        a=a+";"+file+";"+nuevo+ "\n"
        print(file)
    except:
        print("No se pudo obtener: " + file)

with open('c:/data/tapas/Corpus Clarin.csv', 'w', newline="\n") as f:
    f.write(a)
f.close
```
